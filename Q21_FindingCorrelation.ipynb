```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame

# Step 1: Ensure the columns are numeric (convert if necessary)
df['Amount_numeric'] = pd.to_numeric(df['Amount_numeric'], errors='coerce')
df['OverallRoleSat'] = pd.to_numeric(df['OverallRoleSat'], errors='coerce')

# Check for missing values after conversion
print(df[['Amount_numeric', 'OverallRoleSat']].info())

# Check the number of unique values in each column
print(f"Unique values in Amount_numeric: {df['Amount_numeric'].nunique()}")
print(f"Unique values in OverallRoleSa: {df['OverallRoleSat'].nunique()}")

# Step 2: Drop rows with NaN values
df.dropna(subset=['Amount_numeric', 'OverallRoleSat'], inplace=True)

# Step 3: Compute Pearson Correlation
pearson_corr = df['Amount_numeric'].corr(df['OverallRoleSat'], method='pearson')

# Step 4: Compute Spearman Correlation
spearman_corr = df['Amount_numeric'].corr(df['OverallRoleSat'], method='spearman')

# Step 5: Print the results
print(f"Pearson Correlation: {pearson_corr}")
print(f"Spearman Correlation: {spearman_corr}")
```

```python
import pandas as pd
import numpy as np
import re



# Step 1: Convert 'WorkExp' ranges to numerical values (average of the range)
def convert_work_exp(work_exp):
    if work_exp == 'Missing':
        return np.nan  # Handle "Missing" by converting to NaN
    match = re.match(r'(\d+)\s*to\s*(\d+)\s*years', work_exp)
    if match:
        start = int(match.group(1))
        end = int(match.group(2))
        return (start + end) / 2  # Return the average of the range
    return np.nan  # In case the format doesn't match

# Apply the conversion function to the 'WorkExp' column
df['WorkExp'] = df['WorkExp'].apply(convert_work_exp)

# Step 2: Handle NaN values in 'WorkExp' and 'Amount_numeric'
df.dropna(subset=['WorkExp', 'Amount_numeric'], inplace=True)  # Drop rows with NaN in these columns

# Step 3: Compute Pearson Correlation
pearson_corr = df['WorkExp'].corr(df['Amount_numeric'], method='pearson')

# Step 4: Compute Spearman Correlation
spearman_corr = df['WorkExp'].corr(df['Amount_numeric'], method='spearman')

# Step 5: Print the results
print(f"Pearson Correlation: {pearson_corr}")
print(f"Spearman Correlation: {spearman_corr}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder


# Step 1: Convert 'Country' column to numeric using Label Encoding
encoder = LabelEncoder()
df['Country_numeric'] = encoder.fit_transform(df['Country'])
df['Industry'] = encoder.fit_transform(df['Industry'])
df['OrgSize'] = encoder.fit_transform(df['OrgSize'])
df['ProfessionalTech'] = encoder.fit_transform(df['ProfessionalTech'])
df['DevType'] = encoder.fit_transform(df['DevType'])
df['RemoteWork'] = encoder.fit_transform(df['RemoteWork'])
df['Employment'] = encoder.fit_transform(df['Employment'])
df['EdLevel'] = encoder.fit_transform(df['EdLevel'])

# Step 2: Calculate the correlation matrix for the relevant columns
corr_matrix = df[['Country_numeric','RemoteWork','EdLevel', 'Employment','OverallRoleSat','YearsCodePro','DevType','ProfessionalTech','OrgSize','JobStTeamDrivingStgy','Industry','Amount_numeric']].corr()

# Step 3: Plot the heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```
